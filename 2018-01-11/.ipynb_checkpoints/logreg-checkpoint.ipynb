```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from pandas.plotting import scatter_matrix
from itertools import combinations
from sklearn import linear_model
from sklearn.model_selection import cross_val_score
%matplotlib inline
```

```python
df = pd.read_csv('loan.csv')
df.dropna(axis=0, inplace=True)
```

```python
df['approved'] = df['Interest.Rate'] <= 12
```

```python
df.head()
```

```python
scatter_matrix(df, figsize=(12,12))
None
```

```python
y = df.approved
X = df[['FICO.Score','Loan.Amount','Loan.Length','Monthly.Income']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
model = LogisticRegression().fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y = df.approved
rows = []
for i in range(1,5):
    combos = list(combinations(['FICO.Score','Loan.Amount','Loan.Length','Monthly.Income'],i))
    for j,com in enumerate(combos):
        X = pd.DataFrame(df, columns=com)
        model = LogisticRegression(fit_intercept=True)
        accuracy = cross_val_score(model, X, y, cv=10, scoring='accuracy').mean()
        s = ', '.join(com)
        rows.append({'Accuracy':accuracy, 'Columns':s})
df1 = pd.DataFrame(rows)
df1
```
